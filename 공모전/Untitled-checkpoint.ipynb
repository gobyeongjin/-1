### 컬럼별 count

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_fire(data, k):
    data = data.groupby(k).size().reset_index(name='count')
    data = data.sort_values(by='count', ascending=False).head(20)
    plt.figure(figsize=(20, 7))
    sns.barplot(x=k, y='count', data=data)
    plt.xlabel(k)
    plt.ylabel('Count')
    plt.grid()
    plt.xticks(rotation=45)
    plt.show()


def plot_fire1(data, k):
    data = data.groupby(k).size().reset_index(name='count')
    data = data.sort_values(by='count', ascending=False).head(80)
    plt.figure(figsize=(20, 7))
    sns.barplot(x=k, y='count', data=data)
    plt.xlabel(k)
    plt.ylabel('Count')
    plt.grid()
    plt.xticks(rotation=45)
    plt.show()


# Assuming fire_weather is your DataFrame
data = fire_weather

columns = [
    '지역명',
    '화재유형 대분류',
    '발화요인 대분류',
    '발화요인 소분류',
    '인명피해소계',
    '사망',
    '부상',
    '재산피해소계',
    '장소대분류',
    '장소중분류',
    '장소소분류'
]


columns1 = [
    '최고기온(℃)',
    '최저기온(℃)',
    '일교차',
    '강수량(mm)',
    '최대풍속(m/s)',
    '평균풍속(m/s)',
    '평균습도(%rh)',
    '최저습도(%rh)',
    '최대순간풍속(m/s)_numeric'
    
]



# 각 컬럼에 대해 그래프 출력
for column in columns:
    plot_fire(data, column)

for column in columns1:
    plot_fire1(data, column)
```

### 화재유형 대분류

```python
# 범주형 열을 더미 변수로 변환합니다.
dummy_columns = ["화재유형 대분류"]
dummy_df = pd.get_dummies(fire_weather[dummy_columns])

# 수치형 열과 결합합니다.
numeric_columns = ['일교차','강수량(mm)', '평균풍속(m/s)',"최저습도(%rh)" ]
combined_df = pd.concat([fire_weather[numeric_columns], dummy_df], axis=1)

# 상관 행렬을 계산합니다.
corr_matrix = combined_df.corr()

# 히트맵을 그립니다.
plt.figure(figsize=(15, 12))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.title('상관 행렬 히트맵')
plt.show()
```

```python
print("일교차")
print("-------------")
print(combined_df.corr()["일교차"].sort_values(ascending=True))
print("_____________________________________________________________________")
print("강수량(mm)")
print("-------------")
print(combined_df.corr()["강수량(mm)"].sort_values(ascending=True))
print("_____________________________________________________________________")
print("평균풍속(m/s)")
print("-------------")
print(combined_df.corr()["평균풍속(m/s)"].sort_values(ascending=True))
print("_____________________________________________________________________")
print("최저습도(%rh")
print("-------------")
print(combined_df.corr()["최저습도(%rh)"].sort_values(ascending=True))
print("_____________________________________________________________________")
```
